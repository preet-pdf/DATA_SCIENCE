```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
import sklearn
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import KBinsDiscretizer
```

```python
dataset = pd.read_csv(R'/content/drive/MyDrive/dataset/winequalityN.csv')
```

```python
print(dataset.head())
print(dataset.info())
print(dataset.describe())
```

```python
dataset.isnull().any()  
```

```python
dataset['fixed acidity'].fillna(dataset['fixed acidity'].mean(),inplace = True)
dataset['volatile acidity'].fillna(dataset['volatile acidity'].mean(),inplace = True)
dataset['citric acid'].fillna(dataset['citric acid'].mean(),inplace = True)
dataset['residual sugar'].fillna(dataset['residual sugar'].mean(),inplace = True)
dataset['chlorides'].fillna(dataset['chlorides'].mean(),inplace = True)
dataset['pH'].fillna(dataset['pH'].mean(),inplace = True)
dataset['sulphates'].fillna(dataset['sulphates'].mean(),inplace = True)
```

```python
dataset.isnull().any()  
```

```python
lb = LabelEncoder()
dataset['type'] = lb.fit_transform(dataset['type'])
```

```python
dataset.head()
```

```python
for_norm=dataset
for_std=dataset
```

```python
sc = StandardScaler()
for_std = sc.fit_transform(for_std)
for_std
```

```python
for_norm=preprocessing.normalize(for_norm)
pd.DataFrame(for_norm,columns=dataset.columns)
```

```python
dis = preprocessing.KBinsDiscretizer(n_bins=8, encode='ordinal',strategy='quantile')
x=dis.fit_transform(dataset)
```

```python
pd.DataFrame(x,columns=dataset.columns)
```

```python
dis = preprocessing.KBinsDiscretizer(n_bins=8, encode='ordinal',strategy='uniform')
x=dis.fit_transform(dataset)
pd.DataFrame(x,columns=dataset.columns)
```
